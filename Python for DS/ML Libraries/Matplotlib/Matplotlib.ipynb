```python
#Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#scatter is used to scatter the points
plt.scatter(x,y,c = 'g') #c = g is used to define the color
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("Graph in 2D")
```
